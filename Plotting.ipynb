```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ballot_df = pd.read_csv("PyPoll/Resources/election_data.csv")
ballot_df.head()
```

```python
grouped_by_candidate_ballot_df = ballot_df.groupby("Candidate").count()
del grouped_by_candidate_ballot_df["County"]
grouped_by_candidate_ballot_df.head()
```

```python
grouped_by_candidate_ballot_df.plot(kind="bar", y="Voter ID", title="Voting Outcome")
plt.ylabel("Number of Votes")
```

```python
grouped_by_candidate_ballot_df.plot(kind="pie", x="Candidate", y="Voter ID", title="Vote Percentages")
```

```python
grouped_by_county_ballot_df = ballot_df.groupby("County").count()
del grouped_by_county_ballot_df["Candidate"]
grouped_by_county_ballot_df.head()
```

```python
grouped_by_county_ballot_df.plot(kind="bar", y="Voter ID", title="County Vote Distribution")
plt.ylabel("Number of Votes")
```

```python
grouped_by_county_and_candidate_ballot_df = ballot_df.groupby(["County", "Candidate"]).count()
grouped_by_county_and_candidate_ballot_df.head(20)
```

```python
Counties = ballot_df["County"].unique()
Counties

for county in Counties:
    grouped_by_county_and_candidate_ballot_df.loc[county,:].plot(kind="bar", title=f"{county} County Voting Outcome")
    plt.ylabel("Number of Votes")
    
    grouped_by_county_and_candidate_ballot_df.loc[county,:].plot(kind="pie", x="Candidate",
                                                                 y="Voter ID", title=f"{county} County Vote Percentages")
```
